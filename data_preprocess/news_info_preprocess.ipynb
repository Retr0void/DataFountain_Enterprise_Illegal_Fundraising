```python
import pandas as pd
import numpy as np
```

## 预处理news_info数据

```python
news_info = pd.read_csv("D:/COURSES/DataScience/competition/raw_data/news_info.csv")
```

```python
news_info
```

- 将新闻舆情转为数值数据，方便表示

```python
news_info_1 = news_info.replace({'积极': 1, '中立': 0, '消极': -1})
```

```python
news_info_1
```

- 对每个企业的新闻舆情进行加权求和，时间越近的权重越大，防止数据波动过大，只取近十年的样本

```python
for i in range(0, len(news_info_1)):
    if news_info_1.loc[i,'public_date'][:4] == '2020' or '分钟' in news_info_1.loc[i,'public_date'] or '小时' in news_info_1.loc[i,'public_date']:
        news_info_1.loc[i,('positive_negtive')] = news_info_1.loc[i,('positive_negtive')] / np.log(2)
    elif news_info_1.loc[i,('public_date')][:4] == '2019':
        news_info_1.loc[i,('positive_negtive')] = news_info_1.loc[i,('positive_negtive')] / np.log(3)
    elif news_info_1.loc[i,('public_date')][:4] == '2018':
        news_info_1.loc[i,('positive_negtive')] = news_info_1.loc[i,('positive_negtive')] / np.log(4)
    elif news_info_1.loc[i,('public_date')][:4] == '2017':
        news_info_1.loc[i,('positive_negtive')] = news_info_1.loc[i,('positive_negtive')] / np.log(5)
    elif news_info_1.loc[i,('public_date')][:4] == '2016':
        news_info_1.loc[i,('positive_negtive')] = news_info_1.loc[i,('positive_negtive')] / np.log(6)
    elif news_info_1.loc[i,('public_date')][:4] == '2015':
        news_info_1.loc[i,('positive_negtive')] = news_info_1.loc[i,('positive_negtive')] / np.log(7)
    elif news_info_1.loc[i,('public_date')][:4] == '2014':
        news_info_1.loc[i,('positive_negtive')] = news_info_1.loc[i,('positive_negtive')] / np.log(8)
    elif news_info_1.loc[i,('public_date')][:4] == '2013':
        news_info_1.loc[i,('positive_negtive')] = news_info_1.loc[i,('positive_negtive')] / np.log(9)
    elif news_info_1.loc[i,('public_date')][:4] == '2012':
        news_info_1.loc[i,('positive_negtive')] = news_info_1.loc[i,('positive_negtive')] / np.log(10)
    elif news_info_1.loc[i,('public_date')][:4] == '2011':
        news_info_1.loc[i,('positive_negtive')] = news_info_1.loc[i,('positive_negtive')] / np.log(11)
```

```python
idx = []
years = ('2010','2009','2008','2007','2006','2005','2004','2003','2002')
for label, content in news_info_1['public_date'].items():
    #print(f'{label}: {content}')
    if content.startswith(years):
        idx.append(label)
len(idx)
```

```python
news_info_1 = news_info_1.drop(index=idx)
```

```python
news_info_1
```

```python
news_info_1['positive_negtive'].describe()
```

- 按照企业id统计

```python
atti_score = news_info_1.groupby(news_info_1['id']).sum()
```

```python
atti_score
```

```python
atti_score.describe()
```

- 归一化前先处理较大的数据和较小的数据，设置阈值-2和2

```python
for i in range(0,len(atti_score)):
    if atti_score.iat[i,0] < -2:
        atti_score.iat[i,0] = -2
    elif atti_score.iat[i,0] > 2:
        atti_score.iat[i,0] = 2
```

```python
atti_score
```

```python
atti_score.describe()
```

- 接下来数据归一化，映射到[-1,1]

```python
mini = atti_score['positive_negtive'].min()
maxi = atti_score['positive_negtive'].max()
```

**注意**：要先保存最大最小值，防止在数据更新过程中min()、max()函数相应的更新导致结果出错

```python
mini
```

```python
maxi
```

```python
# 归一系数
k = 2 / (maxi - mini)

for i in range(0,len(atti_score)):
    atti_score.iat[i,0] = k * (atti_score.iat[i,0] - mini) -1
```

```python
atti_score
```

```python
atti_score.describe()
```
